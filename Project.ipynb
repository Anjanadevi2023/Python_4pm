```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('https://www.worldometers.info/coronavirus/#countries', 'html.parser')
soup =requests.get(url)
soup =BeautifulSoup(soup.text, "lxml")
soup
```

```python
final_data = []
tr_tags = table_code.find_all('tr')
for i in tr_tags:
    x = i.text.split('\n')
    if x[1] != "":
        final_data.append(x[1:-1])
#final_data    
```

```python
import csv
file = open('covid_data.csv', 'w')
x = csv.writer(file)
x.writerows(final_data)
```
